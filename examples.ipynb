```python
#No. of channels from color image
import numpy

img=cv2.imread("blue flower.jpg")
print('No of Channels is:'+str(img.ndim))
print('No of Channels is:',img.shape)
cv2.imshow("Channel",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#No. of channels from grey-scale image
import numpy

img=cv2.imread("blue flower.jpg",0)
print('No of Channels is:'+str(img.ndim))
print('No of Channels is:',img.shape)

cv2.imshow("Channel",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#binary image
import cv2
img=cv2.imread('flower 1.jpg',0)
ret,bw_img=cv2.threshold(img,127,255,cv2.THRESH_BINARY)

#converting to its binary form
bw=cv2.threshold(img,127,255,cv2.THRESH_BINARY)

cv2.imshow("Binary",bw_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
img=cv2.imread('flower 1.jpg')
B,G,R=cv2.split(img)
print(B)
print(G)
print(R)
```

```python
#blue channel
cv2.imshow("Blue",B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#green channel
cv2.imshow("Green",G)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Red channel
cv2.imshow("Red",R)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
